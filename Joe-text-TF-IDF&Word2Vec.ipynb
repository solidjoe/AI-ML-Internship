```python
import nltk
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('omw-1.4', quiet=True)
nltk.download('punkt_tab', quiet=True)
```

```python
text = """In many traditions of thought, and in many quiet reflections of the human mind, people have asked a question that is both simple and endlessly deep: what does art point to, and why does it exist at all? When we stand before a painting, or when we listen to a piece of music, or when we read a poem that somehow touches the heart, we often feel as if, perhaps, there is something more than just color, sound, or language there. It is not only paint on a canvas; it is not only notes arranged in time; it is not merely words, commas, pauses, and punctuation marks. Rather, it seems to suggest, gently and persistently, that behind the work there is a mind, a will, a creator."""
```

```python
text2 = """Now, of course, the conversation does not end there. Some might argue that art is only human expression, nothing more; it is merely the outcome of neurons firing, of culture evolving, of materials interacting. But even if that were so, the question returns again, almost stubbornly: why does the human mind itself seek meaning? Why do we, again and again, create symbols, stories, images, and sounds that reach beyond survival, beyond utility, beyond mere function? The existence of art, in this sense, raises a further question about the nature of the human person, and perhaps, quietly, about the nature of reality itself."""
```

```python
#Single input
text_combined = text + "\n" + text2

#Removing Punctuations and Tokenisation
text_clean = re.sub(f"[{re.escape(string.punctuation)}]", "", text_combined)
tokens = nltk.word_tokenize(text_clean.lower())

#Stop word removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

final_clean_text = " ".join(filtered_tokens)

print("Tokens after Punctuation & Stop Word Removal:\n", filtered_tokens[:10])
print(final_clean_text)
```

```python
#Lemmatization

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_output = [stemmer.stem(word) for word in filtered_tokens]
lemmatized_output = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print(f"{'Original':<15} | {'Stemmed':<15} | {'Lemmatized'}")
print("-" * 45)
for i in range(15):
    print(f"{filtered_tokens[i]:<15} | {stemmed_output[i]:<15} | {lemmatized_output[i]}")
```

```python
print(final_clean_text)
```

```python
#Keep the final text from the previous assignment, and run the above program to get to this.
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd

input_data = [final_clean_text]

# Bag of Words
bow_vectorizer = CountVectorizer()
bow_matrix = bow_vectorizer.fit_transform(input_data)

# Convert to a Dataframe
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=bow_vectorizer.get_feature_names_out())
print("Bag of Words Result")
print(bow_df.iloc[:, :20])

# Term Frequency and Inverse Document Freq
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(input_data)

# Convert to Dataframe
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("TF-IDF Result ")
print(tfidf_df.iloc[:, :30])
```

```python
!pip install gensim
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

long_text = """Charles Babbage KH FRS (/ˈbæbɪdʒ/; 26 December 1791 – 18 October 1871) was an English polymath.[1] A mathematician, philosopher, inventor and mechanical engineer, Babbage originated the concept of a digital programmable computer.[2]

Babbage is considered by some to merit the title of "father of the computer".[2][3][4][5] He is credited with inventing the first mechanical computer, the difference engine, that eventually led to more complex electronic designs, though all the essential ideas of modern computers are to be found in his analytical engine, programmed using a principle openly borrowed from the Jacquard loom.[2][6] As part of his computer work, he also designed the first computer printers.[7] He had a broad range of interests in addition to his work on computers, covered in his 1832 book Economy of Manufactures and Machinery.[8] He was an important figure in the social scene in London, and is credited with importing the "scientific soirée" from France with his well-attended Saturday evening soirées.[9][10] His varied work in other fields has led him to be described as "pre-eminent" among the many polymaths of his century.[1]

Babbage, who died before the complete successful engineering of many of his designs, including his Difference Engine and Analytical Engine, remained a prominent figure in the ideating of computing. Parts of his incomplete mechanisms are on display in the Science Museum in London. In 1991, a functioning difference engine was constructed from the original plans. Built to tolerances achievable in the 19th century, the success of the finished engine indicated that Babbage's machine would have worked."""

processed_text = word_tokenize(long_text.lower())

data = [processed_text]

model = Word2Vec(data, min_count=1, vector_size=100, window=5, sg=0)

# Check for similarity scores

word1 = "the"
word2 = "computer"
score = model.wv.similarity(word1, word2)

print(f"Word2Vec Similarity: ")
print(f"Similarity score between '{word1}' and '{word2}': {score:.4f}")

print(f"\nWords most similar to 'computer':")
print(model.wv.most_similar("computer", topn=3))
```
