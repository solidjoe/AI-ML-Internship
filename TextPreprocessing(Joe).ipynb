```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('omw-1.4', quiet=True)
```

```python
text = """In many traditions of thought, and in many quiet reflections of the human mind, people have asked a question that is both simple and endlessly deep: what does art point to, and why does it exist at all? When we stand before a painting, or when we listen to a piece of music, or when we read a poem that somehow touches the heart, we often feel as if, perhaps, there is something more than just color, sound, or language there. It is not only paint on a canvas; it is not only notes arranged in time; it is not merely words, commas, pauses, and punctuation marks. Rather, it seems to suggest, gently and persistently, that behind the work there is a mind, a will, a creator."""
```

```python
text2 = """Now, of course, the conversation does not end there. Some might argue that art is only human expression, nothing more; it is merely the outcome of neurons firing, of culture evolving, of materials interacting. But even if that were so, the question returns again, almost stubbornly: why does the human mind itself seek meaning? Why do we, again and again, create symbols, stories, images, and sounds that reach beyond survival, beyond utility, beyond mere function? The existence of art, in this sense, raises a further question about the nature of the human person, and perhaps, quietly, about the nature of reality itself."""
```

```python
#Single input
text_combined = text + "\n" + text2

#Removing Punctuations and Tokenisation
text_clean = re.sub(f"[{re.escape(string.punctuation)}]", "", text_combined)
tokens = nltk.word_tokenize(text_clean.lower())

#Stop word removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

final_clean_text = " ".join(filtered_tokens)

print("Tokens after Punctuation & Stop Word Removal:\n", filtered_tokens[:10])
print(final_clean_text)
```

```python
#Lemmatization

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_output = [stemmer.stem(word) for word in filtered_tokens]
lemmatized_output = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print(f"{'Original':<15} | {'Stemmed':<15} | {'Lemmatized'}")
print("-" * 45)
for i in range(15):
    print(f"{filtered_tokens[i]:<15} | {stemmed_output[i]:<15} | {lemmatized_output[i]}")
```
