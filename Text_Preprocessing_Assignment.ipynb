```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import string

# Download required data
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')

# PARAGRAPH 1 - WITH MULTIPLE PUNCTUATION MARKS
paragraph1 = """
Cloud computing has revolutionized how businesses manage their IT infrastructure!
Instead of maintaining expensive on-premise servers, companies can now use cloud
services provided by vendors such as Amazon Web Services (AWS), Microsoft Azure,
and Google Cloud Platform. Cloud computing offers several advantages: scalability,
cost-efficiency, and flexibility. The three main service models are Infrastructure
as a Service (IaaS)-which provides virtualized computing resources over the internet;
Platform as a Service (PaaS)-which offers development environments and tools; and
Software as a Service (SaaS)-which delivers applications directly to users. Organizations
ask: "How can we migrate to the cloud?" The answer requires careful planning, but the
benefits are significant-reduced capital expenditure, improved reliability, and enhanced
security. Many enterprises have already transitioned to cloud-based solutions!
"""

# PARAGRAPH 2 - WITH MULTIPLE PUNCTUATION MARKS
paragraph2 = """
Amazon Web Services (AWS) is the leading cloud platform, offering a comprehensive suite
of services. Popular AWS services include EC2 (Elastic Compute Cloud) for virtual servers,
S3 (Simple Storage Service) for scalable object storage, RDS (Relational Database Service)
for managed databases, and Lambda for serverless computing. AWS also provides services for
networking (VPC, Route 53), analytics (Redshift, Athena), machine learning (SageMaker), and
security (IAM, KMS). Companies using AWS can: reduce operational costs, improve application
performance, and scale resources on-demand-all without managing physical hardware! However,
organizations must understand AWS pricing models; they include on-demand instances, reserved
instances, and spot instances. Professionals certified in AWS (Solutions Architect, Developer
Associate) are in high demand. The cloud computing industry continues to grow rapidly-with
projections showing substantial increases in market share over the coming years!
"""

# COMBINE 2 PARAGRAPHS AS 1 INPUT
combined_text = paragraph1 + " " + paragraph2

print("\n" + "="*80)
print("TEXT PREPROCESSING ASSIGNMENT")
print("="*80)

print("\n--- ORIGINAL TEXT (2 PARAGRAPHS ON CLOUD COMPUTING & AWS WITH MULTIPLE PUNCTUATION) ---")
print(combined_text.strip())

# Step 1: Remove punctuation marks
print("\n" + "="*80)
print("STEP 1: REMOVE PUNCTUATION MARKS")
print("="*80)
print(f"\nPunctuation marks to be removed: {string.punctuation}")
print("These include: . , : ; ! ? ( ) - \" ' [ ] { } / @ # $ % ^ & * + = ~ ` |")

text_without_punct = combined_text.translate(str.maketrans('', '', string.punctuation))
print("\nText after removing punctuation:")
print(text_without_punct.strip())

# Step 2: Tokenization
print("\n" + "="*80)
print("STEP 2: TOKENIZATION")
print("="*80)

tokens = text_without_punct.lower().split()
print(f"\nTotal tokens: {len(tokens)}")
print(f"Tokens: {tokens}")

# Step 3: Remove stop words
print("\n" + "="*80)
print("STEP 3: STOP WORD REMOVAL")
print("="*80)

stop_words = set(stopwords.words('english'))
print(f"\nTotal stop words: {len(stop_words)}")

filtered_tokens = [word for word in tokens if word not in stop_words]
print(f"\nTokens after removing stop words: {len(filtered_tokens)}")
print(f"Filtered tokens: {filtered_tokens}")

# Step 4: Stemming using Porter Stemmer
print("\n" + "="*80)
print("STEP 4: STEMMING (PORTER STEMMER)")
print("="*80)

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]

print(f"\nStemmed tokens: {stemmed_tokens}")

print("\nStemming examples (Original -> Stemmed):")
print("-" * 60)
for original, stemmed in zip(filtered_tokens, stemmed_tokens):
    if original != stemmed:
        print(f"  {original:20} -> {stemmed}")

# Step 5: Lemmatization
print("\n" + "="*80)
print("STEP 5: LEMMATIZATION")
print("="*80)

lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print(f"\nLemmatized tokens: {lemmatized_tokens}")

print("\nLemmatization examples (Original -> Lemmatized):")
print("-" * 60)
for original, lemmatized in zip(filtered_tokens, lemmatized_tokens):
    if original != lemmatized:
        print(f"  {original:20} -> {lemmatized}")

# Summary
print("\n" + "="*80)
print("SUMMARY STATISTICS")
print("="*80)
print(f"\nOriginal tokens: {len(tokens)}")
print(f"After stop word removal: {len(filtered_tokens)}")
print(f"After stemming: {len(stemmed_tokens)}")
print(f"After lemmatization: {len(lemmatized_tokens)}")
print(f"Reduction rate: {((len(tokens) - len(filtered_tokens)) / len(tokens)) * 100:.2f}%")

# Final Output
print("\n" + "="*80)
print("FINAL PREPROCESSED TEXT")
print("="*80)

print("\nStemmed version:")
print(" ".join(stemmed_tokens))

print("\n\nLemmatized version:")
print(" ".join(lemmatized_tokens))
```
